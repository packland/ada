#### Sistemas Lineares  
Conjunto de equacões lineares, exemplo com $m$ equacões e n incógnitas 

Lembra que estudamos sobre **SPD, SPI e SI**:

**Sistema Possível e Determinado (SPD)**: Apenas uma solucão, quando o determinante é diferente de zero.  
**Sistema Possível e Indeterminado (SPI)**: Solucões possíveis infinítas, quando o determinante é zero.  
**Sistema Impossível (SI)**: Não é possível encontrar solucão, determinante principal é zero e o secundário é diferente de zero.  
  
Regra de Cramer:  
https://pt.wikipedia.org/wiki/Regra_de_Cramer

Regra de Sarrus:  
https://pt.wikipedia.org/wiki/Regra_de_Sarrus

Vamos resolver utilizando um funcão do numpy, lembrando que é para sistemas lineares com solucão única: 

$Ax = b$

```python
import numpy as np
```

```python
A = np.array([
    [1, 2],
    [3, 4]
])
A
```

```python
b = np.array([1,0])
b
```

```python
np.linalg.det(A)
```

```python
x = np.linalg.solve(A, b)
x
```

##### Caso Gráfico

```python
vendas = np.random.randint(50, 100, size=12)
vendas
```

```python
media_vendas = np.mean(vendas)
media_vendas
```

```python
meses = np.arange(1, 13)
meses
```

```python
print(vendas)
tendencia_vendas = np.diff(vendas)
tendencia_vendas
```

```python
tendencia_vendas = np.insert(tendencia_vendas, 0, 0)
tendencia_vendas
```

```python
# https://matplotlib.org/stable/api/markers_api.html
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axhline.html
```

```python
dados_vendas = np.concatenate([vendas, tendencia_vendas])

passo_y = 10

menor_y = np.min(dados_vendas) - passo_y
maior_y = np.max(dados_vendas) + passo_y

print(f'Limites {menor_y} - {maior_y}')

grade_y = np.arange(start=menor_y, stop=maior_y, step=passo_y)
grade_y
```

```python
plt.plot(meses, vendas, marker='o', label='Vendas Mensais')
plt.plot(meses, tendencia_vendas, marker='.', color='g', label='Vendas Tendência')
plt.axhline(y=media_vendas, color='r', linestyle='--', 
            label=f'Vendas Média {media_vendas:.1f}')

plt.xlabel('Meses')
plt.ylabel('Quantidade Vendas (un)')

plt.xticks(meses)
plt.yticks(grade_y)
plt.grid(True)
plt.legend()
plt.show()
```

##### Caso Log

```python
log_data = [
    '2024-01-01 08:00:00 - Processo iniciado',
    '2024-01-01 08:00:00 - Processo Finalizado',
]
```

```python
log_data_estruturado = np.empty(len(log_data), dtype=[('timestamp', 'U19'), ('evento', 'U50')])

for i, linha in enumerate(log_data):
    timestamp, evento = linha.split(' - ')
    log_data_estruturado[i] = (timestamp, evento)

log_data_estruturado

np.savetxt('log_estruturado.csv', 
           log_data_estruturado, 
           delimiter=',', 
           header='timestamp, evento', fmt='%s')
```

##### Trabalhando com imagens
- Data augmentation: processo de criar novas imagens para treino de Deeplearning. Variações das imagens: rotação, escala, translação, mudança de cores e etc...
- Criar metadados dos arquivos de imagem

https://scikit-image.org/

```python
from skimage import io
from scipy import ndimage
```

```python
def carregar_imagem_teste():
    return io.imread('imagem_teste.png')
```

```python
imagem = carregar_imagem_teste()
```

```python
def mostrar_imagem(imagem, titulo=''):
    plt.title(titulo)
    plt.imshow(imagem)
    plt.show()
```

```python
mostrar_imagem(imagem)
```

```python
imagem_rotacionada = ndimage.rotate(imagem, 90)

mostrar_imagem(imagem_rotacionada, 'Rotação 90')
```

```python
imagem_flip = np.fliplr(imagem)
mostrar_imagem(imagem_flip, 'Flip')
```

```python
imagem_deslocada = ndimage.shift(imagem,(500, 0, 0))
mostrar_imagem(imagem_deslocada, 'Deslocada')
```

```python
imagem_zoom = ndimage.zoom(imagem, (0.1, 1, 1))
mostrar_imagem(imagem_zoom)
```

# Introdução ao Numpy

Numpy é uma biblioteca das principais bibliotecas do Python, devido a sua grande velocidade e eficiência ele é extremamente utilizada para computação científica e analise de dados. Esta fama se da principalmente a estrutura chamada [numpy array](https://docs.scipy.org/doc/numpy/reference/arrays.html), um forma eficiente de guardar e manipular matrizes, que serve como base para as tabelas que iremos utilizar.
- Devem ser **homogêneas**
- **É possível** aplicar uma **operação** de forma **direta**, ex. $[1, 2] . 3 > [3, 6]$  
- Elementos são **armazenados** de forma **contínua na memória**  
- Consome **menos memória**  
- **Rápida** para comparar elementos  
- **Muitas funções** para operar  
- É possível **integrar** com códigos em C/C++ e Fortran  
- Recursos para **Algebra Linear** e **Transformada de Fourier**

E as listas?  https://www.geeksforgeeks.org/python-lists-vs-numpy-arrays/  
- Podem ser **heterogênas**  
- Mutável  
- Permite duplicados  
- **Não é possível** aplicar uma **operação** de forma **direta**, ex $[1, 2] x 3 > [1, 2, 1, 2, 1, 2]$  
- Consome **mais memória**  
- Mais **lenta** para comparar elementos  
- **Poucas funções** de operação  
- Elementos **não** são armazenados de forma **contínua na memória**

### Combinando
  
`np.concatenate((a, d), axis=0)`  # Concatenate arrays  
`np.vstack((a, b))`  #Stack arrays vertically(rowwise)  
`np.hstack((e, f))`  #Stack arrays horizontally(columnwise)  

```python
matriz_inicial = np.arange(1, 5)
matriz_final = np.arange(5, 9)
print(matriz_inicial, matriz_final)
```

```python
np.concatenate([matriz_inicial, matriz_final])
```

```python
matriz_rescalada = matriz_inicial.reshape([2, 2])
print(matriz_rescalada)

matriz_rescalada.T
```

```python
np.vstack([matriz_inicial, matriz_final])
```

```python
np.vstack([matriz_inicial.reshape([2, 2]), matriz_final.reshape([2, 2])])
```

```python
np.hstack([matriz_inicial.reshape([2, 2]), matriz_final.reshape([2, 2])])
```

```python
# Bibliotecas

import pandas as pd
import random
from parse import parse
```

```python
#functions

def do_extract_sample(arquivo_entrada, arquivo_saida, recorte):
    with open(arquivo_entrada, 'r') as arquivo_grande:
        linhas = arquivo_grande.readlines()
        total_linhas = len(linhas)
        tamanho_amostra = int(total_linhas * recorte)
        amostra = random.sample(linhas, tamanho_amostra)
        with open(arquivo_saida, 'w') as arquivo_amostra:
            arquivo_amostra.writelines(amostra)
    return True
```

```python
# Configuracoes

arquivo_original = 'arquivos\\access.log'
```

```python
#separar o log em uma amostra menor pra fazer os testes
nome_saida = 'arquivos\\amostra.log'
percentual = 0.01

do_extract_sample(arquivo_original, nome_saida, percentual)


```
